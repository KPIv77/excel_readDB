```python
import pandas as pd
import sqlite3
```

```python
# Add sheet focus
# GSM, UMTS, LTE , NR
```

```python
# Select sheet name in Excel file and set name .DB file.
sheetName = "sheet target"
name_DB = "name file DB"
```

```python
# Add path excel file to read (file in folder).
Df = pd.read_excel(f'path file excel convert to DB', sheet_name=sheetName, header=None)
```

```python
# Set the first row as header
Df.columns = Df.iloc[1]
```

```python
# Connect to SQLite database
conn = sqlite3.connect(f"folder target\\{name_DB}.db")
```

```python
# Write DataFrame to SQL table
Df.to_sql(f"{name_DB}", conn, index=False, if_exists="replace")
```

```python
# Close the connection
conn.close()
```
