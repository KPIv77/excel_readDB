```python
import pandas as pd
import sqlite3
```

```python
# Add sheet focus
```

```python
# Select sheet name in Excel file and set name .DB file.
sheetName = "iris_data"
name_DB = "iris_data"
```

```python
# Add path excel file to read (file in folder).
Df = pd.read_excel(f'/home/kph/project/excel_readDB/file_exportDB/iris_example_data.xlsx', sheet_name=sheetName, header=None)
```

```python
# Set the first row as header
Df.columns = Df.iloc[0]
```

```python
# Connect to SQLite database
conn = sqlite3.connect(f"/home/kph/project/excel_readDB/file_read/{name_DB}.db")
```

```python
# Write DataFrame to SQL table
Df.to_sql(f"{name_DB}", conn, index=False, if_exists="replace")
```
